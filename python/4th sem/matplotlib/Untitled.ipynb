# line


```python
import numpy as np
import matplotlib.pyplot as p
a=np.array([5,10,15,20])
p.plot(a)
p.show()
```

```python
import numpy as np
import matplotlib.pyplot as p
a=np.array([5,10,15,20])
p.plot(a,color="hotpink")
p.show()
```

```python
import numpy as np
import matplotlib.pyplot as p
a=np.array([5,10,15,20])
p.plot(a,color="hotpink",linewidth="10")
p.show()
```

# type of lines

```python
import numpy as np
import matplotlib.pyplot as p
a=np.array([5,10,15,20])
p.plot(a,linestyle="dotted")#ls=":"
p.show()
```

```python
import numpy as np
import matplotlib.pyplot as p
a=np.array([5,10,15,20])
p.plot(a,ls="--")#--=dashed ,:=dotted ,-=soilid , -.=dashdoted
p.show()
p.plot(a,ls="-.")
p.show()
#p.plot(a,ls=".-") this shows error in reverse manner
p.plot(a,"or")
```

# MULTIPLE LINES

```python
import sys
import matplotlib
matplotlib.use('Agg')
import numpy as np
import matplotlib.pyplot as p
a=np.array([5,10,15,20])
b=np.array([5,10,15,20])
p.plot(a)
p.plot(b)
p.show()
p.savefig(sys.stdout.buffer) 
sys.stdout.flush()
```
